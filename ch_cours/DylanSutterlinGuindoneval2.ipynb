        <li> fonction 1 doit renvoyer le troisième pandas.DataFrame </li>
    </ul></ol></div>

```python
import pandas
def make_df(columns, index):
    data = {a : [str(a) + str(i) for i in index] for a in columns}
    return pandas.DataFrame(data, index)
def fonction1(columns, index):
    df2 = make_df(columns, index)
    #créer un deuxième df pandas
    df_copy = df2.copy
    #concaténation des deux jeux de donnés pour former le 3e
    combinaison =  df2.append(df_copy, ignore_index=True)
    return combinaison
```
