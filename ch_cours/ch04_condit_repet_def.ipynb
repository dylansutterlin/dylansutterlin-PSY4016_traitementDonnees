```python
# objectif: Activer la déclaration "if"

py = "thon"

if py == "thon":
    print('oui')
elif len(py) > 2:
    text = 'py'+py
    print(text)
else:
    print('erreur')
```

```python
import pandas
#bouvle intérieure gauche --> recoit l'index pour les row
#dans la plus grande bloucle extérieure stack des clés 
def make_df(columns, index):
    ##prend la clé a + le nb de i ==> e.g. A1
    data = {a: [str(a) + str(i) for i in index] for a in columns}
    return pandas.DataFrame(data, index)

##***premier 'argument' = colonne et le deuxième ligne
##c avant l donc colonne et ensuite ligne
#a = colonne
#Index 
```

<div class="alert alert-success">
<ul>
    <li>Lorsque le nombre des variables d'un argument n'est pas connu, le symbole étoile * peut être utilisé avant le paramètre

```python
#l'étoile indique que le paramètre est un tuple
def fonc_args(*params):
    print(params[2])

fonc_args('nom1', 'nom2', 'nom3', 'nom4')
```

```python
def fonc_args(*args):
    print(args[2])
    
fonc_args('nom1', 'nom2', 'nom3', 'nom4')
```

```python
v1 = "nom1"
v2 = "nom2"
v3 = 'nom3'

fonc_args(v1, v2, v3)
```

<div class="alert alert-success">
<ul>
    <li> Lorsque le nombre d'arguments n'est pas connu, ajoutez deux astérisques: ** avant le nom du paramètre.
    <li> la fonction recevra un DICTIONNAIRE d'arguments
    <li> ce paramètre est appelé **kwargs 
    </ul></div>

```python
#au lieu de créer un tuple avec une seule étoile, ici les 2 étoiles = dict
def fonc_kwargs(**kwargs):
    print(kwargs['var2'])

fonc_kwargs(var1 = 'nom1', var2 = 'nom2')
```

<div class="alert alert-success">
<ul>
    <li>le **kwargs n'est pas un argument
    </ul></div>

```python
var1 = 'nom1'
var2 = 'nom2'

fonc_kwargs(var1: 'lol', var2: 'lol2')
```

```python
def make_df(columns, index):
    """Cette fonction crée un cadre de données
    Args:
        columns = chaîne de lettres (str)
        index   = liste avec des nombres (int)
    Return:
        pandas.DataFrame
    """
    data = dict()
    for a in columns:
        data[a] = list()
    for key in data:
        for i in index:
            data[key].append(key+str(i))
    return pandas.DataFrame(data, index)
```
